```python
import pandas as pd
from matplotlib import pyplot as pl
from matplotlib.offsetbox import AnchoredText
import cartopy.crs as ccrs
from mpl_toolkits.basemap import Basemap
import cartopy.feature as cfeature
import numpy as np
```

```python
dataset = pd.read_csv('IRNAID5100.csv')
print(dataset.shape)

# Filter to stations only in USA
dataset = dataset.loc[dataset['Group Name'].str.contains('USA')]
print(dataset.shape)
```

```python
dataset.columns
```

```python
# Prep the location information of all 70 cm repeaters
## 420MHz to 450MHz

band_dataset = dataset.loc[(dataset['Frequency'] > 420) & (dataset['Frequency'] < 450)]

proj = ccrs.PlateCarree()

ax = pl.axes(projection=proj)
ax.set_extent([-128,-63,50,22], crs=proj)
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)

fig = ax.get_figure()
fig.set_size_inches(15,15)

ax.coastlines()
#ax.add_feature(cfeature.NaturalEarthFeature('physical', 'land', '50m', edgecolor='face', facecolor=cfeature.COLORS['land']))
ax.add_feature(cfeature.NaturalEarthFeature('physical', 'ocean', '50m', edgecolor='face', facecolor=cfeature.COLORS['water']))

ax.add_feature(cfeature.STATES, edgecolor='grey')
ax.add_feature(cfeature.BORDERS,edgecolor='black')

m = Basemap(projection='cyl', llcrnrlat=20, urcrnrlat=50, llcrnrlon=-130, urcrnrlon=-60)
m.scatter(band_dataset['Longitude'], band_dataset['Latitude'], marker='x', color='k')

ax.set_title('70 cm Repeater Stations')
```
